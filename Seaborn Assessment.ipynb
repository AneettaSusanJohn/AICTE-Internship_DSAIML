```
# 1. Load the 'tips' dataset from Seaborn and display the first 5 rows.

import seaborn as sns
import matplotlib.pyplot as plt
tips = sns.load_dataset('tips')
print(tips.head())
```

```
# 2. Create a scatter plot of total_bill vs tip using the 'tips' dataset.

sns.scatterplot(x='total_bill', y='tip', data=tips)
plt.title('Scatter Plot of Total Bill vs Tip')
plt.xlabel('Total Bill ($)')
plt.ylabel('Tip ($)')
plt.show()
```

```
# 3. Plot a line graph for total_bill vs tip using Seaborn.

sns.lineplot(x='total_bill', y='tip', data=tips)
plt.title('Line Plot of Total Bill vs Tip')
plt.xlabel('Total Bill ($)')
plt.ylabel('Tip ($)')
plt.show()
```

```
# 4. Draw a bar plot to show the average tip for each day in the 'tips' dataset.


sns.barplot(x='day', y='tip', data=tips, estimator='mean')
plt.title('Average Tip per Day')
plt.xlabel('Day')
plt.ylabel('Average Tip ($)')
plt.show()
```

```
# 5. Use a count plot to display how many records are available for each day.

sns.countplot(x='day', data=tips)
plt.title('Number of Records per Day')
plt.xlabel('Day')
plt.ylabel('Count')
plt.show()
```

```
# 6. Make a box plot to show total_bill distribution across different days.

sns.boxplot(x='day', y='total_bill', data=tips)
plt.title('Total Bill Distribution Across Days')
plt.xlabel('Day')
plt.ylabel('Total Bill')
plt.show()
```

```
# 7. Create a violin plot for total_bill vs day.

sns.violinplot(x='day', y='total_bill', data=tips)
plt.title('Violin Plot of Total Bill by Day')
plt.xlabel('Day')
plt.ylabel('Total Bill')
plt.show()
```

```
# 8. Plot a histogram of the 'total_bill' column with KDE using histplot.

sns.histplot(data=tips, x='total_bill', kde=True)
plt.title('Histogram and KDE of Total Bill')
plt.xlabel('Total Bill')
plt.ylabel('Frequency')
plt.show()
```

```
# 9. Load the 'iris' dataset and draw a pairplot for all features.

iris = sns.load_dataset('iris')
sns.pairplot(iris)
plt.show()

```

```
# 10. Use the 'titanic' dataset and show a heatmap of its numeric correlation matrix.

titanic = sns.load_dataset('titanic')

corr_matrix = titanic.corr(numeric_only=True)
plt.figure(figsize=(10, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', linewidths=0.5)

plt.title('Correlation Heatmap of Titanic Dataset')
plt.show()
```

```
# 11. Use the 'penguins' dataset and draw a scatter plot of bill_length_mm vs flipper_length_mm.


penguins = sns.load_dataset('penguins')

plt.figure(figsize=(8,6))
sns.scatterplot(data=penguins, x='bill_length_mm', y='flipper_length_mm')

plt.title('Scatter plot of Bill Length vs Flipper Length')
plt.xlabel('Bill Length (mm)')
plt.ylabel('Flipper Length (mm)')
plt.show()
```

```
# 12. Plot a box plot for 'fare' against 'class' using the 'titanic' dataset.

plt.figure(figsize=(8,6))
sns.boxplot(data=titanic, x='class', y='fare')

plt.title('Box Plot of Fare by Passenger Class')
plt.xlabel('Class')
plt.ylabel('Fare')
plt.show()
```

```
# 13. Create a count plot to show the number of survivors and non-survivors in the Titanic dataset.

plt.figure(figsize=(6,5))
sns.countplot(data=titanic, x='survived')

plt.title('Count of Survivors and Non-Survivors')
plt.xlabel('Survived (0 = No, 1 = Yes)')
plt.ylabel('Count')
plt.show()
```

```
# 15. Create a heatmap of the correlation matrix from the 'penguins' dataset using numeric_only=True.

corr = penguins.corr(numeric_only=True)

plt.figure(figsize=(8,6))
sns.heatmap(corr, annot=True, cmap='coolwarm')

plt.title('Correlation Matrix - Penguins Dataset')
plt.show()

```

```
# 16. Load the 'diamonds' dataset and draw a violin plot of price by cut.

diamonds = sns.load_dataset('diamonds')

plt.figure(figsize=(10,6))
sns.violinplot(data=diamonds, x='cut', y='price')

plt.title('Violin Plot of Diamond Price by Cut')
plt.xlabel('Cut')
plt.ylabel('Price')
plt.show()
```

```
# 17. Draw a pairplot using the 'penguins' dataset with hue as 'species'.

sns.pairplot(data=penguins)

plt.suptitle('Pairplot of Penguins Dataset by Species', y=1.02)
plt.show()
```

```
# 19. Apply a dark theme to your plots using set_style and recreate a scatter plot.

sns.set_style("dark")

plt.figure(figsize=(8, 6))
sns.scatterplot(data=penguins, x="bill_length_mm", y="bill_depth_mm", hue="species")

plt.title("Penguins: Bill Length vs Bill Depth (Dark Theme)", color='white')

plt.gca().set_facecolor("#2c2c2c")

plt.show()
```

```
# 20. Convert all categorical columns in the 'tips' dataset to numeric and create a full correlation heatmap.

import pandas as pd

tips_numeric = pd.get_dummies(tips, drop_first=True)

corr = tips_numeric.corr()

plt.figure(figsize=(12, 8))
sns.heatmap(corr, annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.5)

plt.title("Correlation Heatmap of Tips Dataset (Categorical Columns Converted)")

plt.show()

```
